```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_hotdog = pd.read_excel("hotdog-contest-winners.xlsm")
```

```python
df_obama = pd.read_excel("obama-approval-ratings.xls")
```

```python
df_hotdog.head()
```

```python
df_obama.head()
```

```python
sns.barplot(x = 'Dogs eaten', y = 'Winner', data=df_hotdog)
plt.title('Hot Dogs Eaten by Winners')
plt.show()
```

```python
df_obama.set_index('Issue').plot(kind = 'barh', stacked=True, color = ['blue', 'red', 'grey'])
```

```python
sns.lineplot(x='Year', y='Dogs eaten', data=df_hotdog)
plt.title('Hot Dogs Eaten by Year')
plt.show()
```

```python
y = np.array([df_obama['Approve'][2], df_obama['Disapprove'][2], df_obama['None'][2]])
plt.pie(y, labels = ['Approve','Disapprove','None'], colors = ['blue', 'red', 'grey'], autopct = '%1.1f%%')
plt.title("Obama's Terrorism Approval Ratings")
plt.show()
```

```python
y = np.array([df_obama['Approve'][9], df_obama['Disapprove'][9], df_obama['None'][9]])
plt.pie(y, labels = ['Approve', 'Disapprove', 'None'], colors = ['blue', 'red', 'grey'], autopct = '%1.1f%%')
circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(circle)
plt.title("Obama's Economy Approval Ratings")
plt.show()
```
